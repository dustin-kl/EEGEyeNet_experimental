```python
from tensorflow import keras
import numpy as np
from keras import backend as K 
```

```python
model = keras.models.load_model('./runs/1625588790_Age_task_antisaccade_min_tensorflow_lr_0.01/checkpoint/run1/CNN_nb_0_')
```

```python
model.summary()
```

```python
new_model = keras.Model(inputs=model.input, outputs=model.layers[-3].output)
```

```python
data = np.load('./data/Age_task_with_antisaccade_synchronised_min.npz')
```

```python
X = data['EEG']
```

```python
X.shape
```

```python
features = new_model(X)
```

```python
features.shape
```

```python
features = K.eval(features)
features.shape
```

```python
np.save('features_antisaccade_age.npy', features)
```
