```python
import numpy as np
import copy 
import sys
from tqdm import tqdm
```

```python
tar_pos = [[400, 300],
 [650, 500],
 [400, 100],
 [100, 450],
 [700, 450],
 [100, 500],
 [200, 350],
 [300, 400],
 [100, 150],
 [150, 500],
 [150, 100],
 [700, 100], 
 [300, 200], 
 [100, 100], 
 [700, 500], 
 [500, 400], 
 [600, 250], 
 [650, 100],
 [200, 250],
 [400, 500],
 [700, 150],
 [500, 200],
 [100, 300],
 [700, 300],
 [600, 350],
]
```

```python
data = np.load('./data/Position_task_with_dots_synchronised_min.npz')
X = data['EEG']
y = data['labels']
```

```python
y_backup = copy.deepcopy(y)		
```

```python
y
```

```python
new_labels = []
text_file = open("./label_log.txt", "w")
for i in tqdm(range(len(y))):
	coord = np.array([y[i][1], y[i][2]])
	min_dist = sys.maxsize
	min_point = None 
	for j in range(len(tar_pos)):
		point = tar_pos[j]
		dist = ((coord - np.array(point))**2).mean()
		if dist < min_dist:
			min_dist = dist
			min_point = j
	label = np.zeros(2)
	label[0] = y[i][0] # participant ID 
	label[1] = min_point

	text_file.write(f"Sample {i}:    coord: {coord}, class: {min_point}, point: {tar_pos[min_point]} \n")

	new_labels.append(label)

text_file.close()
```

```python
np_labels = np.array(new_labels)
np.max(np_labels[:,1])
```

```python
np.savez("./data/Dots_Classification_task_with_dots_synchronised_min.npz", EEG=X, labels=np_labels)
```
